```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../panda/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
bnb_neighborhood_groups = air_bnb.groupby('neighbourhood_group').size().sort_values(ascending=False)
display(bnb_neighborhood_groups)

print(f'The neighborhood which shows up the most is Manhattan')
```

```python
# Are private rooms the most popular in manhattan?
air_bnb_parsed = air_bnb[['id','neighbourhood_group','room_type']]
pop_room_type = air_bnb_parsed.loc[air_bnb['neighbourhood_group'] == 'Manhattan'].groupby(['room_type']).size()
print(pop_room_type)
print('\nThe most popular room type in mahattan is: Entire home/apt')
```

```python
# Which hosts are the busiest and based on their reviews?
top_hosts = air_bnb.groupby('host_id')['number_of_reviews'].size().sort_values(ascending=False).reset_index()
display(top_hosts.head(10))

id = top_hosts['host_id'].iloc[0]
print(f'The busiest host based on their reviews is host id {id}')
```

```python
#Which neighorhood group has the highest average price?
ave_price = air_bnb.groupby('neighbourhood_group')['price'].mean().round().sort_values(ascending=False)
display(ave_price.to_frame())

print('The neighborhood group with the highest average price is Manhattan')
```

```python
# Which neighbor hood group has the highest total price?
neighborhood_max_price = air_bnb.groupby('neighbourhood_group')['price'].sum().round().sort_values(ascending=False)
neighborhood_max_price.to_frame()
```

```python
#Which top 5 hosts have the highest total price?
# top_hosts.head(5)

hosts_merged = air_bnb.merge(top_hosts, on = 'host_id', how = 'outer')[['host_id', 'neighbourhood_group', 'price']]
hosts_merged_sorted = hosts_merged.groupby('host_id')['price'].sum().round().sort_values(ascending=False)

print('The Top 5 Hosts based on reviews and their total price:')
hosts_merged_sorted.to_frame().head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
sorted_air_bnb = air_bnb.sort_values('host_name', kind = 'mergesort')[['host_name','number_of_reviews','availability_365']]
top_review = sorted_air_bnb.loc[sorted_air_bnb['number_of_reviews'] >= 100][['host_name','number_of_reviews']]
# display(top_review)
zero_available = sorted_air_bnb.loc[sorted_air_bnb['availability_365'] == 0][['host_name','availability_365']]
# display(zero_available)

print('Hosts with 100 reviewsor more and no availibility:')
top_non_available = top_review.merge(zero_available, on = 'host_name', how = 'inner').drop_duplicates(subset = 'host_name')
display(top_non_available.head(20))
```

```python
# What host has the highest total of prices and where are they located?
expensive_host = air_bnb.groupby('host_name')['price'].sum()
expensive_host_df = expensive_host.to_frame()
# display(expensive_host_df.head(10))
host_info = air_bnb[['host_id','host_name','neighbourhood']]
# host_info.head(10)
expensive_host_info = expensive_host_df.merge(host_info, on = 'host_name', how = 'outer').sort_values('price', ascending=False)
expensive_host_info.head(10)

most_expensive_host = expensive_host_info.iloc[0]
host = most_expensive_host['host_name']
location = most_expensive_host['neighbourhood']
print(f'The host with the highest total of prices is {host} and the are located in {location}')
```

```python
# When did Danielle from Queens last receive a review?
last_review = air_bnb.loc[air_bnb['host_name'] == 'Danielle']['last_review'].sort_values(ascending=False).dropna().to_frame()
danielle_from_queens = last_review.loc[air_bnb['neighbourhood_group'] == 'Queens']
# display(danielle_from_queens)

date = danielle_from_queens['last_review'].iloc[0]
print(f'Danielle from Queens recieved their last review on {date}.')
```

## Further Questions

1. Which host has the most listings?

```python
most_listings = air_bnb.groupby('host_name')['calculated_host_listings_count'].count().sort_values(ascending=False).to_frame()
display(most_listings.head(10))
```

2. How many listings have completely open availability?

```python
open_availability = air_bnb.loc[air_bnb['availability_365'] == 365][['id','host_name','availability_365']]
open_availability.count().to_frame()
```

3. What room_types have the highest review numbers?

```python
room_by_reviews = air_bnb.groupby('room_type')['number_of_reviews'].mean().round().sort_values(ascending = False)
display(room_by_reviews.to_frame())
```
